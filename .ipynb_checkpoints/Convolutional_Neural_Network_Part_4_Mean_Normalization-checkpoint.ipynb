```python
# importing libraries
import seaborn as sns
import pickle as pkl
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# inputting sample into numpy variable
input_ = np.append(pkl.load(open(r'C:\Users\abelp\machine_learning\crowd_count\data\final_input_aj_1.pkl', 'rb')),
                  pkl.load(open(r'C:\Users\abelp\machine_learning\crowd_count\data\final_input_aj_2.pkl', 'rb')), axis=0)
```

```python
# print out example frame
plt.imshow(input_[100])
plt.title('Sample Frame')
plt.savefig(r'C:\Users\abelp\machine_learning\crowd_count\data_preprocessing\sample_frame1.png')
plt.show()
```

```python
# taking the mean image
mean_value = input_[0:2000][:,:,:].mean(axis=0)
```

```python
# show the mean image
plt.imshow(mean_value)
plt.title('Mean Frame')
plt.savefig(r'C:\Users\abelp\machine_learning\crowd_count\data_preprocessing\mean_frame.png')
plt.show()
```

```python
# show the mean image subtracted from example frame
plt.imshow(input_[100] - mean_value)
plt.title("Subtracting the mean from sample frame")
plt.savefig(r'C:\Users\abelp\machine_learning\crowd_count\data_preprocessing\subtracted_mean_frame.png')
plt.show()
```

```python
# print out the absolute value of the mean subtracted from the sample frame
plt.imshow(abs(input_[100] - mean_value))
plt.title("Absolute value of subtracting the mean from sample frame")
plt.savefig(r'C:\Users\abelp\machine_learning\crowd_count\data_preprocessing\abs_subtracted_mean_frame.png')
plt.show()
```

```python
# normalizing the samples
mean_normalized = abs(input_ - mean_value)
```

```python
# Create pickle files of the normalized images
pkl.dump(mean_normalized[0:1000], open(r'C:\Users\abelp\machine_learning\crowd_count\final_data\mean_normalized1.pkl', 'wb'))
pkl.dump(mean_normalized[1000:], open(r'C:\Users\abelp\machine_learning\crowd_count\final_data\mean_normalized2.pkl', 'wb'))
```
